```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
world.head()
world.plot(column='pop_est', scheme='quantiles');
```

```python
world2 = world[['continent', 'geometry', 'pop_est']]
continents = world2.dissolve(by='continent', aggfunc='sum')
continents.plot(column = 'pop_est', scheme='quantiles', cmap='YlOrRd');
continents.head()
```
